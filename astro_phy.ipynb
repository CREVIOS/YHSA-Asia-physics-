<marquee><h1><b>The discrete Fourier transform</marquee></h1></b>


First, let's define the discrete Fourier transform (DFT, discrete Fourier transform). Let be a time series $ u [j] = u (t_ {j}) $ with $ t_ {j} = j 
\Delta t = j T / N $ where $ T $ is the recording time, $ N $ le number of measurements and $ j = 0,1...., N-1 $ the index of time measurement points. We will decompose this signal on a finite number of frequencies defined as $ f_ {n} = n / T $ with $ n = 0,1,...N - 1 $ the frequency index.
The DFT is then defined as follows:



$$ \hat{u}[n] = \frac{1}{N} \sum_{j=0}^{N-1}u[j] e^{-2 i \pi \frac{nj}{N}}$$

The inverse Fourier transform, i.e. the operation which reconstructs the time signal from the frequency spectrum, is thus

$$u[j] = \sum_{j=0}^{N-1} \hat{u}[n] e^{2 i \pi \frac{nj}{N}}$$

Such a transform has interesting properties: for a real signal, it is possible to show that:

$$ \hat{u}[N - n] = \hat{u}^*[n]$$
***Proof :***

$$ \hat{u}[N - n] = \frac{1}{N} \sum_{j=0}^{N-1}u[j] e^{-2 i \pi \frac{(N-n)j}{N}} = \frac{1}{N} \sum_{j=0}^{N-1}u[j] e^{-2i \pi j} e^{2 i \pi \frac{nj}{N}} = \frac{1}{N} \sum_{j=0}^{N-1}u[j] e^{2 i \pi \frac{nj}{N}} = \hat{u}^*[n]$$

We can also demonstrate that this transformation conserves energy:

$$ \sum_{n=0}^{N-1} \left| \hat{u}[n] \right| ^2  = \sum_{n=0}^{N-1} \hat{u}[n] \hat{u}[n]^* = \frac{1}{N^2} \sum_{n = 0}^{N-1} \left( \sum_{j=0}^{N-1} u[j] e^{-2i\pi \frac{nj}{N}} \right) \left( \sum_{k=0}^{N-1} u[k] e^{2i\pi \frac{nk}{N}} \right)$$
By separating the product of sums into a sum of sums, it comes:

$$ \sum_{n=0}^{N-1} \left| \hat{u}[n] \right| ^2 = \frac{1}{N^2} \sum_{n = 0}^{N-1} \left( \sum_{\substack{j=0\\ k=0\\ j=k}}^{N-1} u[j] u[k] + \sum_{\substack{j=0\\ k=0\\ j\neq k}}^{N-1} u[j]u[k] e^{2i\pi \frac{n(k-j)}{N}} \right)$$
Reversing the summations lets you write

$$ \sum_{n=0}^{N-1} \left| \hat{u}[n] \right| ^2 = \frac{1}{N^2} \sum_{j=0}^{N-1} u[j]^2 \sum_{n=0}^{N-1} 1 + \sum_{\substack{j=0\\ k=0\\ j\neq k}}^{N-1} u[j]u[k] \sum_{n=0}^{N-1} e^{2i\pi \frac{n(k-j)}{N}}$$
where we recognize the sum of the roots of unity which is equal to zero. From where :
$$ \sum_{n=0}^{N-1} \left| \hat{u}[n] \right| ^2 = \frac{1}{N} \sum_{j=0}^{N-1} \left| u[j]\right|^2$$

This is the Parseval-Plancherel theorem.

We talked above about the PSD (power spectral density). It is defined: $ S [n] = 2T \left | \hat {u} [n] \right | $ with $ n = 0,1,\ldots, N / 2-1 $, because the second half of the spectrum is identical to the first (as we showed earlier) . $ S [n] $ is expressed in $ \text {V} ^ 2 ~ \text {Hz} ^ {- 1} $. We can also rewrite Plancherel's theorem with the PSD:

$$ \sum_{n=0}^{N-1} S[n] = \frac{2T}{N} \sum_{j=0}^{N-1} \left| u[j]\right|^2$$

**Morlet wavelet transform**

Let Morlet's wavelet define as follows with respect to time $ t $: $$ \psi_0 (t) = \pi ^{- 1/4} e ^{- i \omega_0 t} e ^ {- t ^ 2/2} $$ The expansion and translation of this wavelet allows to define by convolution the transform into wavelets: $$ \mathcal {W} (\tau, t) = \sum_ {j = 0} ^ {N- 1} u (t_j) \psi ^ * ((t_j - t) / \tau) $$ with $ t_j $ the position around which we convolve in a window $ \Delta t $, and $ \tau $ un time dilation parameter or time scale. We choose this scale so that a simple relation exists with the Fourier frequencies: $ 1 /f \simeq \tau $ by setting $ \omega_0 = 6 $.
The square of the modulus of the wavelet coefficient $ \left | \mathcal {W} (\tau, t) \right | ^ 2 $ representing the quantum of energy of the fluctuations of $ u (t) $ on the surface $ \Delta T \times \Delta \tau $ around a moment $ t $ on the scale $ \tau $, we can plot the scalogram of the transform on a two-dimensional graph $ (t, \tau) $. It is also possible to go back to the spectral power equivalent to the PSD via the spectrum of fluctuations by integrating on the time variable:

$$ S_w = 2 \delta t \mathcal {W} (f) ^ 2 = \frac {2 \delta t} {N} \sum_ {i = 0} ^ {N-1} \left | \mathcal {W} (f, t_j) \right | ^ 2 $$

```

N = int(1e3)
T = 10
A0 = 10
f0 = 0.5
dt = T/N
A1 = 8
f1 = 1

t = np.arange(N) * dt
y = y0(t, A0, f0)
plt.plot(t, y, color='red', linewidth=1.3)
plt.xlabel(r't (s)')
plt.ylabel(r'$y_{0}$')
plt.xlim([np.min(t), np.max(t)])
plt.show()
```
